```python
import gym
import matplotlib.pyplot as plt

# 1) Specify render_mode when you make the env:
env = gym.make("MountainCar-v0", render_mode="rgb_array")

# 2) Reset returns (observation, info) in Gymnasium; if you’re on Gym <=0.25 it’s just obs:
obs, info = env.reset()

# 3) Now render() will return a frame array:
frame = env.render()  

plt.figure()
plt.imshow(frame)
plt.axis('off')
plt.show()
```
